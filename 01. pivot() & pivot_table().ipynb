```python
# import pandas library for data manipulation and analysis
import pandas as pd
```

## **pivot()**

```python
# read and show the dataset
df_gdp = pd.read_csv('gdp.csv', encoding='latin-1')
df_gdp
```

```python
# reshape the dataframe with .pivot()
df_gdp_pivot = df_gdp.pivot(index='year', columns='country', values='gdppc')
df_gdp_pivot
```

## **pivot_table()**

```python
# read and show the dataset
df_sales = pd.read_excel('supermarket_sales.xlsx')
df_sales
```

```python
# make a pivot table and add an aggregate function
df_sales.pivot_table(
    index='Gender',
    values=['Quantity', 'Rating', 'Tax 5%', 'Total', 'Unit price', 'cogs', 'gross income', 'gross margin percentage'],
    aggfunc='sum'
).round(1)
```

```python
# make a pivot table, add an aggregate function and select some columns
df_sales.pivot_table(
    index='Gender',
    values=['Quantity', 'Total'],
    aggfunc='sum'
).round(1)
```

```python
# make a pivot tables that says how much male and female spend in each category
df_sales.pivot_table(
    index='Gender',
    columns='Product line',
    values='Total',
    aggfunc='sum'
).round(1)
```
