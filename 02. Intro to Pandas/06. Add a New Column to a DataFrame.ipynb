## **Adding a New Column**

**Adding a New Column with a Scalar Value**

```python
# adding new column to a dataframe
df_exams['language score'] = 70
df_exams
```

**Adding a New Column with an Array** -> The dataframe has 1000 rows, so we need to create an array with 1000 elements

```python
# import numpy
import numpy as np
```

```python
# create an array of 1000 elements
language_score = np.arange(0, 1000)
```

```python
# length of the array
len(language_score)
```

```python
# adding a new column to a dataframe with an array
df_exams['language score'] = language_score
df_exams
```

```python
# create random integer numbers between 1 and 100
int_language_score = np.random.randint(1, 100, size=1000)
```

```python
# min value inclusive & high value exclusive
print(min(int_language_score))
print(max(int_language_score))
```

```python
# adding a new column to a dataframe with an array
df_exams['language score'] = int_language_score
df_exams
```
