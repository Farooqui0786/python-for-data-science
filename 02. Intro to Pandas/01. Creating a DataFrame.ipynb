```python
# import pandas and numpy libraries for data handling and numerical operations
import pandas as pd
import numpy as np
```

## **Creating a DataFrame from an array**

**Option 1.**

```python
# creating an array
data = np.array([[1, 2], [3, 4], [5, 6]])
```

```python
# creating a dataframe
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'], columns=['col1', 'col2'])
```

```python
# showing the dataframe
df
```

**Option 2.**

```python
# creating an array with list shape
list_data = [[1, 4], [2, 5], [3, 6]]
```

```python
# creating a dataframe
df = pd.DataFrame(list_data, index=['row1', 'row2', 'row3'], columns=['col1', 'col2'])
```

```python
# showing the dataframe
df
```

## **Creating a DataFrame from a Dictionary**

```python
# lists used for this example
states = ["California", "Texas", "Florida", "New York"]
population = [39613493, 29730311, 21944577, 19299981]
```

```python
# storing list within dictionaries
dict_series = {"States": states, "Population": population}
```

```python
# creating a dataframe
df_population = pd.DataFrame(dict_series)
# df_population = pd.DataFrame.from_dict(dict_series)
```

```python
# showing the dataframe
df_population
```
