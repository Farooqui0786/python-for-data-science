```python
# import pandas library for data handling and analysis
import pandas as pd
```

```python
# reading the csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
```

```python
# showing the dataframe
df_exams
```

## **Create Index**

```python
# Import numpy for numerical operations and random for generating random numbers
import numpy as np
import random
```

```python
# creating non-repetitive values for the index
new_index = np.arange(0, 1000)
new_index
```

```python
# shuffling the indexes
np.random.shuffle(new_index)
new_index
```

```python
# create a new column with new index
df_exams['new_index'] = new_index
df_exams
```

## **Set Index**

```python
# set new_index column as new index
df_exams.set_index('new_index', inplace=True)
df_exams
```

## **Sort Index**

```python
# sort dataframe by new index
df_exams.sort_index()
```

```python
# sort descending by index
df_exams.sort_index(ascending=False)
```
