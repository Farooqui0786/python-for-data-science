```python
# import pandas library for data handling and analysis
import pandas as pd
```

```python
# reading the csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
```

```python
# showing the dataframe
df_exams
```

## **Math Operations**

### **Operations in Column**

```python
# select a column and calculate a total sum
df_exams['math score'].sum()
```

```python
# calculate count, mean, std, max, min
print(df_exams['math score'].count())
print(df_exams['math score'].mean())
print(df_exams['math score'].std())
print(df_exams['math score'].max())
print(df_exams['math score'].min())
```

```python
# easier calculation with .describe()
df_exams.describe()
```

### **Operations in Row**

```python
# calculate the sum in a row
df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
```

```python
# calculating the average score and assigning the result to a new column
df_exams['average'] = (df_exams['math score'] + df_exams['reading score'] + df_exams['writing score'])/3
```
