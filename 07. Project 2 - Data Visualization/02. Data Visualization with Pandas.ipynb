```python
# import pandas library for data manipulation and analysis
import pandas as pd
```

```python
# reading the csv file
df_population_raw = pd.read_csv('population_total.csv')
```

```python
# first look at the dataset
df_population_raw
```

## **Making a Pivot Table**

.pivot(): Returns reshaped DataFrame organized by given index/column values ("pivot without aggregation")

```python
# dropping null values
df_population_raw.dropna(inplace=True)
```

```python
# making a pivot table
df_pivot = df_population_raw.pivot(index='year', columns='country', values='population')
```

```python
# selecting some countries
df_pivot = df_pivot[['India', 'China', 'United States', 'Indonesia', 'Brazil']]
```

```python
# showing pivot table
df_pivot
```

## **Line Plot**

```python
# making lineplot
df_pivot.plot(kind='line',
              xlabel='Years', ylabel='Population',
              title='Population (1855-2020)',
              figsize=(12, 6))
```

## **Barplot**

### **Single Barplot**

```python
# selecting only one year
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
# showing new dataframe
df_pivot_2020
```

```python
# transpose dataframe from rows to columns or viceversa
df_pivot_2020 = df_pivot_2020.T
```

```python
# making barplot
df_pivot_2020.plot(kind='bar',
                   xlabel='Years', ylabel='Population',
                   title='Population in 2020',
                   figsize=(12, 6))
```

### **Barplot Grouped by "n" Variables**

```python
# selecting a few years
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
```

```python
# showing new dataframe
df_pivot_sample
```

```python
# making grouped barplot
df_pivot_sample.plot(kind='bar',
                     xlabel='Years', ylabel='Population',
                     title='Population (1980-2020)',
                     figsize=(12, 6))
```

## **Piechart**

```python
# changing column name
df_pivot_2020.rename(columns={2020: '2020'}, inplace=True)
```

```python
# making piechart
df_pivot_2020.plot(kind='pie',
                   y='2020',
                   title='Population in 2020 (%)',
                   figsize=(12, 6))
```

## **Boxplot**

### **Single Boxplot**

```python
# making single boxplot
df_pivot['United States'].plot(kind='box', color='green',
                               ylabel='Population',
                               figsize=(12, 6))
```

### **Multiple Boxplot**

```python
# making multiple boxplot
df_pivot.plot(kind='box', color='green',
              xlabel='Countries', ylabel='Population',
              figsize=(12, 6))
```

## **Histogram**

```python
# making histogram
df_pivot[['United States', 'Indonesia']].plot(kind='hist',
                                              xlabel='Population',
                                              figsize=(12, 6))
```

## **Scatterplot**

```python
# selecting a sample of the raw data
df_sample = df_population_raw[df_population_raw['country'].isin(['United States', 'India', 'China', 'Indonesia', 'Brazil'])]
```

```python
# making scatterplot
df_sample.plot(kind='scatter', x='year', y='population',
               color='green', s=80,
               figsize=(12, 6))
```
