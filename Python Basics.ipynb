### **Set Operations & Methods**

**1. Adding elements to a set**

```python
fruits = {"apple", "banana"}
fruits.add("cherry")      # add single element
print(fruits)
print("----")

fruits.update(["mango", "orange"])      # add multiple elements
print(fruits)
```

**2. Removing elements**

```python
fruits = {"apple", "banana", "cherry"}
fruits.remove("banana")     # remove specific item
print(fruits)
print("----")

fruits.discard("mango")     # removes if exists, else does nothing
print(fruits)
print("----")

removed_item = fruits.pop()     # removes a random element
print(removed_item)
print(fruits)     # remaining elements
```

**3. Clearing a set**

```python
fruits = {"apple", "banana", "cherry"}
fruits.clear()      # remove all elements
print(fruits)
```

**4. Set Operations**

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Union
print(set1 | set2)
print("----")

# Intersection
print(set1 & set2)
print("----")

# Difference
print(set1 - set2)
print("----")

# Symmetric difference
print(set1 ^ set2)
```

**5. Updating sets with Operations**

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set1.update(set2)     # adds all elements from set2
print(set1)
print("----")

set1.intersection_update({2, 3, 4})     # keep only common elements
print(set1)
print("----")

set1.difference_update({3})     # remove elements present in another set
print(set1)
```

**6. Copying a set**

```python
fruits = {"apple", "banana", "cherry"}

# 1. Using copy() → creates independent copy
new_set = fruits.copy()
print(new_set)
print("----")

# 2. Using set() constructor
new_set2 = set(fruits)
print(new_set2)
print("----")

# 3. Using = → reference (not independent)
new_set3 = fruits
new_set3.add("mango")
print(fruits)     # original also changes!
```

## **List**

A list in Python is an ordered, mutable collection of items.
* Ordered: Items have a defined position (index).
* Mutable: Items can be changed, added, or removed.
* Can store multiple datatypes in one list.

```python
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4]
mixed = [1, "apple", True, 3.14]
```

**Indexing:** Indexing in a list means accessing items using their position number (index).

```python
fruits = ["apple", "banana", "cherry", "orange"]

# Positive indexing
print(fruits[0])      # 'apple'  (first item)
print(fruits[2])      # 'cherry' (third item)
print("----")

# Negative indexing
print(fruits[-1])     # 'orange' (last item)
print(fruits[-2])     # 'cherry' (second last item)
```

**Slicing:** Slicing means extracting a portion (sub-list) of elements from a list using indexes.

list[start : stop : step]
* start → index where slicing begins (default = 0)
* stop → index where slicing ends (excluded)
* step → how many steps to jump (default = 1)

```python
numbers = [10, 20, 30, 40, 50, 60, 70]

# Basic slicing
print(numbers[1:4])     # [20, 30, 40] → from index 1 to 3
print(numbers[:3])      # [10, 20, 30] → from start to index 2
print(numbers[3:])      # [40, 50, 60, 70] → from index 3 to end
print("----")

# Using step
print(numbers[::2])     # [10, 30, 50, 70] → every 2nd element
print(numbers[1::2])      # [20, 40, 60] → every 2nd element starting from index 1
print("----")

# Negative indexing
print(numbers[-4:-1])     # [40, 50, 60] → slice from -4 to -2
print(numbers[::-1])      # [70, 60, 50, 40, 30, 20, 10] → reversed lis
```

### **List Operations & Methods**

**1. Adding element to a list**

```python
# 1. append() → add item at the end
fruits = ["apple", "banana"]
fruits.append("cherry")
print(fruits)
print("----")

# 2. insert() → add item at specific index
fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)
print("----")

# 3. extend() → add multiple items
fruits = ["apple", "banana"]
fruits.extend(["cherry", "mango"])
print(fruits)
print("----")

# 4. Using = operator (+) → concatenate lists
fruits = ["apple", "banana"]
fruits = fruits + ["grape", "kiwi"]
print(fruits)
print("----")

# 5. Nested list → adding a list inside another list
fruits = ["apple", "banana"]
fruits.append(["cherry", "mango"])
print(fruits)
```

**2. Remove an element**

```python
# 1. remove() → removes first matching element
fruits = ["apple", "banana", "cherry", "banana"]
fruits.remove("banana")
print(fruits)     # only first 'banana' removed
print("----")

# 2. pop(index) → removes element at given index (default = last item)
fruits = ["apple", "banana", "cherry"]
fruits.pop(1)
print(fruits)
print("----")

# 3. del → delete item by index or entire list
fruits = ["apple", "banana", "cherry"]
del fruits[0]
print(fruits)
print("----")

# 4. clear() → removes all elements (makes list empty)
fruits = ["apple", "banana", "cherry"]
fruits.clear()
print(fruits)
```

**3. Sorting a list**

```python
# 1. sort() → sorts list in ascending order (changes original list)
numbers = [5, 2, 9, 1, 7]
numbers.sort()
print(numbers)
print("----")

# 2. sort(reverse=True) → sorts in descending order
numbers = [5, 2, 9, 1, 7]
numbers.sort(reverse=True)
print(numbers)
print("----")

# 3. sorted() → returns a new sorted list (original unchanged)
numbers = [5, 2, 9, 1, 7]
new_list = sorted(numbers)
print(new_list)
print(numbers)      # old unchanged list
print("----")

# 4. Sorting strings (alphabetical order)
fruits = ["banana", "apple", "cherry"]
fruits.sort()
print(fruits)
print("----")

# 5. Custom sorting using key (by length of string)
fruits = ["banana", "apple", "cherry"]
fruits.sort(key=len)
print(fruits)     # sorted by word length
```

**4. Update value on a list**

```python
# Updating by index
fruits = ["apple", "banana", "cherry"]
fruits[1] = "orange"      # change 'banana' → 'orange'
print(fruits)
print("----")

# Updating multiple values using slicing
numbers = [10, 20, 30, 40, 50]
numbers[1:3] = [25, 35]   # replace index 1 and 2
print(numbers)
print("----")

# Replacing a range with more items
numbers[2:4] = [100, 200, 300]
print(numbers)
```

**5. Copying a list**

```python
# 1. Using copy() → makes a shallow copy
fruits = ["apple", "banana", "cherry"]
new_list = fruits.copy()
print(new_list)
print("----")

# 2. Using list() constructor
fruits = ["apple", "banana", "cherry"]
new_list = list(fruits)
print(new_list)
print("----")

# 3. Using slicing [:]
fruits = ["apple", "banana", "cherry"]
new_list = fruits[:]
print(new_list)
print("----")

# 4. Assignment (=) → NOT a real copy (both point to same list)
fruits = ["apple", "banana", "cherry"]
new_list = fruits
new_list[0] = "grape"
print(fruits)     # original also changes!
```

## **Dictionary**

A dictionary in Python is an unordered collection of key-value pairs.
* Key: Unique identifier to access the value.
* Value: Data associated with the key (can be any datatype).
* Dictionaries are mutable → you can add, update, or remove items.

```python
# Creating a dictionary
student = {
    "name": "Muzammil",     # key: name, value: Muzammil
    "age": 20,      # key: age, value: 20
    "course": "Python"      # key: course, value: Python
}

# Accessing value by key
print(student["name"])
print("----")

# Getting all keys
print(student.keys())
print("----")

# Getting all values
print(student.values())
print("----")

# Getting all key-value pairs
print(student.items())
```

### **Dictionary Operations & Methods**

**1. Adding/Updating elements**

```python
# Creating a dictionary
student = {
    "name": "Muzammil",
    "age": 20,
    "course": "Python"
}

# 1. Adding a new key-value pair
student["grade"] = "A"
print(student)
print("----")

# 2. Updating an existing value
student["age"] = 21
print(student)
print("----")

# 3. Using update() to add or change multiple elements
student.update({"course": "Data Science", "city": "Delhi"})
print(student)
```

**2. Removing elements**

```python
# Creating a dictionary
student = {
    "name": "Muzammil",
    "age": 21,
    "course": "Data Science",
    "grade": "A",
    "city": "Delhi"
}

# 1. Using del → remove a specific key-value pair
del student["city"]
print(student)
print("----")

# 2. Using pop() → removes a key and returns its value
removed_value = student.pop("grade")
print(removed_value)      # 'A'
print(student)
print("----")

# 3. Using popitem() → removes the last inserted key-value pair
student.popitem()
print(student)
print("----")

# 4. Using clear() → removes all items, makes dictionary empty
student.clear()
print(student)
```

**3. Copying dictionary**

```python
# Original dictionary
student = {
    "name": "Muzammil",
    "age": 20,
    "course": "Python"
}

# 1. Using copy() → creates a shallow copy
new_dict = student.copy()
print(new_dict)
print("----")

# 2. Using dict() constructor
new_dict2 = dict(student)
print(new_dict2)
print("----")

# 3. Using = → NOT a real copy (both point to same dictionary)
new_dict3 = student
new_dict3["age"] = 21
print(student)      # original unchanged
```

## **Tuple**

A tuple in Python is an ordered, immutable collection of items.
* Ordered: Items have a defined position (index).
* Immutable: Cannot be changed after creation (no adding, removing, or updating).
* Can store multiple datatypes.
* Can be nested (tuple inside tuple).

```python
# Creating a tuple
fruits = ("apple", "banana", "cherry")
numbers = (1, 2, 3, 4)
mixed = (1, "apple", True, 3.14)

# Accessing elements
print(fruits[0])
print(fruits[-1])
print("----")

# Slicing
print(numbers[1:3])
print("----")

# Nested tuple
nested = (1, 2, (3, 4))
print(nested[2])
print(nested[2][0])
```

**Indexing:**

```python
fruits = ("apple", "banana", "cherry", "orange")

# Positive indexing
print(fruits[0])      # 'apple'   (first element)
print(fruits[2])      # 'cherry'  (third element)
print("----")

# Negative indexing
print(fruits[-1])     # 'orange'  (last element)
print(fruits[-2])     # 'cherry'  (second last element)
```

**Slicing:**

```python
numbers = (10, 20, 30, 40, 50, 60)

# Basic slicing
print(numbers[1:4])     # (20, 30, 40) → from index 1 to 3
print(numbers[:3])      # (10, 20, 30) → start to index 2
print(numbers[3:])      # (40, 50, 60) → from index 3 to end
print("----")

# Using step
print(numbers[::2])     # (10, 30, 50) → every 2nd element
print(numbers[1::2])      # (20, 40, 60) → every 2nd element starting from index 1
print("----")

# Negative indexing
print(numbers[-4:-1])     # (30, 40, 50) → slice from -4 to -2
print(numbers[::-1])      # (60, 50, 40, 30, 20, 10) → reversed tuple
```

### **Tuple Operrations & Methods**

**1. Adding elements**

```python
# Original tuple
t = (1, 2, 3)
print("Original tuple:", t)
print("----")

# 1. Using concatenation (+)
t1 = t + (4,)     # add single element
t1 = t1 + (5, 6)      # add multiple elements
print("After concatenation:", t1)
print("----")

# 2. Convert to list → add → back to tuple
temp = list(t)      # convert to list
temp.append(4)      # add single element
temp.extend([5, 6])     # add multiple elements
t2 = tuple(temp)      # convert back to tuple
print("After converting to list and adding:", t2)
print("----")

# 3. Using multiplication (*) → repeats elements
t3 = t * 3
print("After multiplication (repetition):", t3)
```

**2. Removing elements:** Tuples are immutable, so you cannot directly remove elements. But you can work around it by creating a new tuple without the element(s) you want to remove.

```python
# Original tuple
t = (1, 2, 3, 4, 5)
print("Original tuple:", t)
print("----")

# 1. Convert to list → remove → back to tuple
temp = list(t)
temp.remove(3)      # remove a single element
t1 = tuple(temp)
print("After removing 3 using list conversion:", t1)
print("----")

# 2. Remove multiple elements using list comprehension
t2 = tuple(x for x in t if x not in (2, 4))     # remove 2 and 4
print("After removing 2 and 4 using comprehension:", t2)
print("----")

# 3. Remove by slicing (exclude some indexes)
t3 = t[:2] + t[3:]      # remove element at index 2 (third element)
print("After removing element at index 2 using slicing:", t3)
```

**3. Updating elements:** Tuples are immutable, so you cannot directly update elements. But you can create a new tuple with the updated values using concatenation, slicing, or conversion to list.

```python
# Original tuple
t = (1, 2, 3, 4, 5)
print("Original tuple:", t)
print("----")

# 1. Convert to list → update → back to tuple
temp = list(t)
temp[2] = 30      # update element at index 2
t1 = tuple(temp)
print("After updating index 2 using list conversion:", t1)
print("----")

# 2. Using slicing + concatenation to update a range
t2 = t[:1] + (20, 30) + t[3:]     # update index 1 and 2
print("After updating indexes 1 and 2 using slicing:", t2)
print("----")

# 3. Using tuple comprehension (create new tuple with conditions)
t3 = tuple(x*10 if x==3 else x for x in t)      # multiply 3 by 10
print("After conditional update using comprehension:", t3)
```

**4. Copying elements:** Tuples are immutable, so copying them is straightforward. You can create a new tuple that points to the same elements (shallow copy).

```python
# Original tuple
t = (1, 2, 3, 4, 5)
print("Original tuple:", t)
print("----")

# 1. Using assignment (=) → reference copy
t1 = t
print("Reference copy:", t1)
print("----")

# 2. Using slicing [:] → creates a shallow copy
t2 = t[:]
print("Shallow copy using slicing:", t2)
print("----")

# 3. Using tuple() constructor → creates a new tuple
t3 = tuple(t)
print("Copy using tuple():", t3)
```

## **If Statement**

An **if** statement in Python is used for decision-making — it allows the program to run a block of code only if a certain condition is true.

```python
x = 10      # You can change this value to test different cases

# 1. Simple if
if x > 5:
    print("x is greater than 5")      # Runs only if condition is True
print("----")

# 2. if-else
if x > 5:
    print("x is greater than 5")      # Runs if condition is True
else:
    print("x is not greater than 5")      # Runs if condition is False
print("----")

# 3. if-elif-else
if x < 5:
    print("x is less than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is greater than 5")      # Runs if no previous condition matched
```

## **For Loop**

A **for** loop in Python is used to repeat a block of code for each item in a sequence (like list, tuple, string, set, dictionary, or range).

```python
# 1. Loop through a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)      # Prints each fruit in the list
print("----")

# 2. Loop through a string
for letter in "hello":
    print(letter)     # Prints each character in the string
print("----")

# 3. Loop using range()
for i in range(5):      # range(5) → 0,1,2,3,4
    print(i)      # Prints numbers 0 to 4
print("----")

# 4. Loop with if condition
for i in range(1, 6):     # Numbers from 1 to 5
    if i % 2 == 0:
        print(i, "is even")
    else:
        print(i, "is odd")
```

### **Enumerate in a For loop**

**enumerate()** in Python is used in a **for loop** to get both the index and the value of items in a sequence.

```python
fruits = ["apple", "banana", "cherry"]

# 1. Enumerating a list (default index starts from 0)
for index, fruit in enumerate(fruits):
    print(index, fruit)     # Prints index and fruit name
print("----")

# 2. Enumerating a list with custom start index
for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)     # Index starts from 1
print("----")

# 3. Enumerating a string
for i, ch in enumerate("hello"):
    print(i, ch)      # Prints index and character
```

### **Looping through dictionary**

```python
# Sample dictionary
student = {"name": "Alice", "age": 21, "grade": "A"}

# 1. Looping through keys
for key in student:
    print(key)      # Prints only keys
print("----")

# 2. Looping through values
for value in student.values():
    print(value)      # Prints only values
print("----")

# 3. Looping through key-value pairs
for key, value in student.items():
    print(key, ":", value)      # Prints both key and value
print("----")

# 4. Looping with enumerate (index + key + value)
for index, (key, value) in enumerate(student.items(), start=1):
    print(index, key, "=", value)     # Prints index + key + value
```

## **While Loop**

A **while** loop in Python is used to repeat a block of code as long as a condition is true.

```python
# 1. Simple while loop
i = 1
while i <= 5:     # Loop runs while i <= 5
    print(i)
    i += 1      # Increment i
print("----")

# 2. Using break
i = 1
while i <= 5:
    if i == 3:
        break     # Exit the loop when i = 3
    print(i)
    i += 1
print("----")

# 3. Using continue
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue      # Skip printing when i = 3
    print(i)
print("----")

# 4. While with else
i = 1
while i <= 3:
    print(i)
    i += 1
else:
    print("Loop finished!")     # Runs when loop ends normally (not by break)
```

## **Break & Continue**

* **Break:** Immediately terminates the loop (for or while), skipping the remaining iterations.
* Use case: When a condition is met and you want to exit the loop early.
---
* **Continue:** Skips the current iteration of the loop and moves to the next iteration.
* Use case: When you want to ignore certain conditions but keep looping.

## **Functions**

A function in Python is a block of reusable code that performs a specific task. Functions help to organize code, reduce repetition, and make it easier to read and maintain.

### **Built-in Functions**

* Built-in functions in Python are predefined functions that are ready to use without importing any module.
* They perform common tasks like mathematical operations, type conversion, sequence handling, and more.

```python
# 1. abs() → absolute value
print(abs(-5))
print("----")

# 2. len() → length of a sequence
print(len("hello"))
print(len([1,2,3,4]))
print("----")

# 3. type() → returns the data type
print(type(10))
print(type([1,2,3]))
print("----")

# 4. min() and max() → smallest and largest element
numbers = [5, 2, 9, 1]
print(min(numbers))
print(max(numbers))
print("----")

# 5. sum() → sum of elements
print(sum(numbers))
print("----")

# 6. round() → rounds a float
print(round(3.14159, 2))
print("----")

# 7. sorted() → returns a sorted list
print(sorted(numbers))
print(sorted(numbers, reverse=True))
print("----")

# 8. str(), int(), float() → type conversion
print(str(100))
print(int("50"))
print(float("3.14"))
print("----")

# 9. range() → generate sequence of numbers
for i in range(3):
    print(i)
print("----")

# 10. enumerate() → get index and value
fruits = ["apple","banana","cherry"]
for index, fruit in enumerate(fruits):
    print(index, fruit)
```

### **User-defined Functions**

* A user-defined function is a function created by the programmer to perform a specific task.
* It helps to organize code, reuse logic, and avoid repetition.

```python
# 1. Simple function without parameters
def greet():
    print("Hello, World!")      # Prints greeting

greet()     # Calling the function
print("----")

# 2. Function with parameters
def add(a, b):
    return a + b      # Returns sum of a and b

result = add(5, 3)
print("Sum:", result)
print("----")

# 3. Function with default parameter
def greet_name(name="Guest"):
    print(f"Hello, {name}!")      # Prints greeting with name

greet_name()      # Hello, Guest!
greet_name("Alice")     # Hello, Alice!
print("----")

# 4. Function returning multiple values
def arithmetic(a, b):
    return a+b, a-b, a*b, a/b     # Sum, difference, product, division

sum_, diff, prod, div = arithmetic(10, 2)
print(sum_, diff, prod, div)
```

## **Modules**

A module in Python is a file containing Python code (functions, variables, classes, etc.) that can be imported and used in other Python programs.
* Helps to organize code and reuse functionality.
* Python has built-in modules and you can also create custom modules.

### **Using Modules**

```python
# 1. Import the whole module
import math     # Import math module
print("Square root of 16:", math.sqrt(16))      # Use sqrt() → 4.0
print("----")

# 2. Import only specific functions
from math import pi, factorial      # Import only pi and factorial
print("Value of pi:", pi)
print("Factorial of 5:", factorial(5))
print("----")

# 3. Rename module while importing
import math as m      # Give math module a short name 'm'
print("Square root of 25 using alias:", m.sqrt(25))
```
