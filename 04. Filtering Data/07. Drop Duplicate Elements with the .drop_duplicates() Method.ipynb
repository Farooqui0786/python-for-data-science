```python
# import pandas library for data manipulation and analysis
import pandas as pd
```

```python
# read the csv file "laptop_price.csv" into a pandas dataframe with latin-1 encoding
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin-1')
```

```python
# display the first 3 rows of the DataFrame
df_laptops.head(3)
```

## **drop_duplicates()**

```python
# drop duplicates in two or more columns
df_laptops.drop_duplicates(['Company'])
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
# sort dataframe ascending by company and price
# cheadpest first and most expensive last
df_laptops = df_laptops.sort_values(by=['Company', 'Price_euros'])
```

```python
# cheapest: keep='first'
df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Price_euros']]
```

```python
# most expensive: keep='last'
df_laptops.drop_duplicates(['Company'], keep='last')[['Company', 'Price_euros']]
```
