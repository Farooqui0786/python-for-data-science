```python
# import pandas library for data manipulation and analysis
import pandas as pd
```

```python
# read the csv file "laptop_price.csv" into a pandas dataframe with latin-1 encoding
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin-1')
```

```python
# display the first 3 rows of the DataFrame
df_laptops.head(3)
```

## **Creating a Conditional Column from 2 Choices: np.where()**

```python
# import numpy library for numerical operations and array handling
import numpy as np
```

```python
# creating an array based on price tiers
np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Inexpensive')
```

```python
# set it to a new column
df_laptops['Price_tier'] = np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Inexpensive')
```

```python
# show first 5 rows
df_laptops.head()
```

```python
# count values in price_tier column
df_laptops.value_counts('Price_tier')
```

```python
# create an array based on screen size
np.where(df_laptops['Inches'] > 15, 'Big', 'Small')
```

```python
# set it to a new column
df_laptops['Screen_size'] = np.where(df_laptops['Inches'] > 15, 'Big', 'Small')
```

```python
# show first 5 columns
df_laptops.head()
```

```python
# count values in screen_size column
df_laptops.value_counts('Screen_size')
```
