```python
# import pandas library for data manipulation and analysis
import pandas as pd
```

```python
# read the csv file "laptop_price.csv" into a pandas dataframe with latin-1 encoding
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin-1')
```

```python
# display the first 3 rows of the DataFrame
df_laptops.head(3)
```

## **isin(): Single Filtering**

```python
# select "Apple" or "HP" laptops
df_laptops['Company'].isin(['Apple', 'HP'])
```

```python
# filter dataframes
df_laptops[df_laptops['Company'].isin(['Apple', 'HP'])]
```

## **isin(): Multiple Filtering**

```python
# find "Notebook" or "Ultrabook" form "Apple" or "HP"
filter1 = df_laptops['TypeName'].isin(['Notebook', 'Ultrabook'])
filter2 = df_laptops['Company'].isin(['Apple', 'HP'])
```

```python
# filter dataframes
df_laptops[filter1 & filter2]
```
