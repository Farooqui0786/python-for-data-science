```python
# import pandas library for data manipulation and analysis
import pandas as pd
```

```python
# read the csv file "players_20.csv" into a pandas DataFrame
df = pd.read_csv('players_20.csv')

# set index
df.set_index('short_name', inplace=True)

# select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# show dataframe
df
```

## **copy()**

### **deep = True**

```python
# deep=True by default (modifications to the data or indices of the copy will not be reflected in the original object)
df_copy = df.copy()
```

```python
# update value in original dataframe
df.loc['L. Messi', 'height_cm'] = 180
```

```python
# copy vs original dataframe (no change reflecteed in copy dataframe)
df
df_copy
```

### **deep = False**

```python
# deep=False (any changes to the data of the original will be reflected in the shallow copy and vice versa)
df_shallow_copy = df.copy(deep=False)
```

```python
# update value in original dataframe
df.loc['Cristiano Ronaldo', 'height_cm'] = 200
```

```python
# shallow copy vs original dataframe (changes reflected in both)
df
df_shallow_copy
```

### **Simple Assignment**

```python
# make a copy
df_new_copy = df
```

```python
# update value in original dataframe
df.loc['Neymar Jr', 'height_cm'] = 190
```

```python
# copy vs original dataframe (changes reflected in both)
df
df_new_copy
```
