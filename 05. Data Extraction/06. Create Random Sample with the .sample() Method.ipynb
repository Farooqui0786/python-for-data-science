```python
# import pandas library for data manipulation and analysis
import pandas as pd
```

```python
# read the csv file "players_20.csv" into a pandas DataFrame
df = pd.read_csv('players_20.csv')

# set index
df.set_index('short_name', inplace=True)

# select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# show dataframe
df
```

## **sample()**

```python
# extract 10 random elements from the "nationality" column
df['nationality'].sample(10, random_state=99)
```

```python
# extract a random 20% sample of the dataframe
df.sample(frac=0.2, random_state=99)
```

```python
# unsample: increase the sampling rate -> frac > 1 (Note: replace parameter has to be True for frac parameter > 1)
df.sample(frac=2, replace=True, random_state=99)
```
