```python
# import pandas library for data manipulation and analysis
import pandas as pd
```

```python
# read the csv file "players_20.csv" into a pandas DataFrame
df = pd.read_csv('players_20.csv')

# set index
df.set_index('short_name', inplace=True)

# select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# show dataframe
df
```

## **Set Value to One Cell**

```python
# update "L. Messi" height
df.loc['L. Messi', 'height_cm'] = 175
```

```python
# show dataframe
df
```

## **Set Value to Entire Column**

```python
# set height of all players to 190
df.loc[:, 'height_cm'] = 190
```

```python
# show dataframe
df
```

## **Set Value to Entire Row**

```python
# get all columns that correspond to player ranked last in fifa
df.iloc[-1, :]
```

```python
# set null values to all columns that corresponds to player ranked last in fifa
import numpy as np
df.iloc[-1, :] = np.nan
```

```python
# show dataframe
df
```

## **Set Value to Multiple Cells**

```python
# set value for all items matching tha list of labels
df.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm']] = 175
```

```python
# show dataframe
df
```

## **Set Values for Rows Matching a Condition**

```python
# set value for rows matching a condition
columns = ['age', 'dob', 'height_cm', 'weight_kg']

df.loc[df['height_cm'] > 180, columns] = 0
```

```python
# show dataframe
df
```
