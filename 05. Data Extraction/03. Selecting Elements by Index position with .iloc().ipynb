```python
# import pandas library for data manipulation and analysis
import pandas as pd
```

```python
# read the csv file "players_20.csv" into a pandas DataFrame
df = pd.read_csv('players_20.csv')

# set index
df.set_index('short_name', inplace=True)

# select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# show dataframe
df
```

## **Selecting with a Single value**

iloc[row_position, column_position]

```python
# get the height of "L. Messi"
df.iloc[0, 3]
```

```python
# get the weight of "Cristiano Ronaldo"
df.iloc[1, 4]
```

```python
# get all rows inside the "height_cm" column
df.iloc[:, 3]
```

```python
# get all columns that correspond to the index "L. Messi"
df.iloc[0, :]
```

## **Selecting with a List of values**

```python
# get all data about "L. Messi" & "Cristiano Ronaldo"
df.iloc[[0, 1]]
```

```python
# get the height of "L. Messi" & "Cristiano Ronaldo"
df.iloc[[0, 1], 3]
```

```python
# get the height & weight of "L. Messi"
df.iloc[0, [3, 4]]
```

```python
# get height & weight of "L. Messi" & "Cristiano Ronaldo"
df.iloc[[0, 1], [3, 4]]
```

## **Selecting a range of Data with a Slice**

start:stop (Note that the start is inlcuded but the stop is excluded)

```python
# slice column labels age/club
# players = ['L. Messi', 'Cristiano Ronaldo']
players = [0, 1]

df.iloc[players, 1:7]     # age:club+1
```

```python
# slice index labels (top1 to top10 player)
# columns = ['age', 'dob', 'height_cm', 'weight_kg']
columns = [1, 2, 3, 4]

# get top1 and top10 player name
df.iloc[0:10, columns]
```

## **Selecting with Conditions**

iloc can't accept a boolean series, but only a boolean list. We have to use the list() function to convert a series into a boolean list

```python
# one condition: select player with above 180cm
# columns = ['age', 'dob', 'height_cm', 'weight_kg']
columns = [1, 2, 3, 4]

df.iloc[list(df['height_cm'] > 180), columns]
```

```python
# multiple conditions; select player with height above 180cm from "Argentina"
df.iloc[list((df['height_cm'] > 180) & (df['nationality'] == 'Argentina')), :]
```
